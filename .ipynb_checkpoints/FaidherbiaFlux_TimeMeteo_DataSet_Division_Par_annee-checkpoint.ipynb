# Traitement du dataset

```python
import numpy as np
import pandas as pd
import datetime
```

```python
data= pd.read_csv("/home/openopse/Documents/OPENOPSE/data/Olivier Roupsard/updated_new/OpenOPSE_Roupsard/OutdataFFlux_OpenOPSE_Soil.csv")
data.head()
```

```python
dates = [str(datetime.datetime.fromtimestamp(x)) for x in data.DateSemih]
data['DateHeure']= dates
```

```python
annees = [x.split('-',1)[0] for x in data.DateHeure]
data['Year']= annees
data=data.astype({'Year': 'int64'})
```

```python
for y in annee:
    save = data[data['Year'] == y]
    name = '/home/openopse/Documents/OPENOPSE/data/Olivier Roupsard/updated_new/OpenOPSE_Roupsard/per Year/'+str(y)+'/OutdataFFlux_OpenOPSE_Soil_'+str(y)+'.csv'
    save.to_csv(name, index=False)
```
