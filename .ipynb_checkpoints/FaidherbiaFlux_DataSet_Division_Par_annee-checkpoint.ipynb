## segmentation du jeu de données

```python
#meteo soil

df_tsoilfs = data[tsoilfs]
df_vwfs = data[vwfs]
df_tsoilwell = data[tsoilwell]
df_vwwell = data[vwwell]

#tree
df_leafRate = data[leafRate]
df_leafAD = data[leafAD]
df_leafAPT = data[leafAPT]
df_dripLLAI = data[dripLLAI]
df_CWNDVI = data[CWNDVI]
df_LaicwAntPlotScale = data[LaicwAntPlotScale]
df_LaiTreecwAntPlotScale = data[LaiTreecwAntPlotScale]
df_woodW = data[woodW]
df_netRadiation = data[netRadiation]
# meteo
df_msSmallAnt = data[msSmallAnt]
df_msTallAnt = data[msTallAnt]
df_fluxTallAnt = data[fluxTallAnt]
```

```python
annee = data['Year'].unique()
annee
```

## sauvegarder les jeux de données segmentés

```python
for y in annee:
    save = data[data['Year'] == y]
    name = '~/Documents/OPENOPSE/data/Olivier Roupsard/updated/dataset/Outdata_Time_Meteo_Flux_Tsoil_TDR_LAI_Puits2_'+str(y)+'.csv'
    save.to_csv(name, index=False)
```

```python
for y in annee:
    
    name = '~/Documents/OPENOPSE/data/Olivier Roupsard/updated/output/'+str(y)
    #meteo soil
    save = df_tsoilfs[df_tsoilfs['Year'] == y]
    save.to_csv(name+'/soil_temp_profile_under_full_sun_0_to_-2m'+str(y)+'.csv', index=False)
    save = df_vwfs[df_vwfs['Year'] == y]
    save.to_csv(name+'/volumetric_water_content_under_full_sun_0_to_-2m'+str(y)+'.csv', index=False)
    save = df_tsoilwell[df_tsoilwell['Year'] == y]
    save.to_csv(name+'/soil_temp_profile_in_the_wells2_-0,2_to_-4,8m'+str(y)+'.csv', index=False)
    save = df_vwwell[df_vwwell['Year'] == y]
    df_vwwell.to_csv(name+'/volumetric_water_content_in_the_wells2_-0,2_to_-4,8m'+str(y)+'.csv', index=False)

    #tree
    save = df_leafRate[df_leafRate['Year'] == y]
    save.to_csv(name+'Leafing_rate'+str(y)+'.csv', index=False)
    save = df_leafAD[df_leafAD['Year'] == y]
    save.to_csv(name+'/leaf_area_density'+str(y)+'.csv', index=False)
    save = df_leafAPT[df_leafAPT['Year'] == y]
    save.to_csv(name+'/leaf_area_per_tree'+str(y)+'.csv', index=False)
    save = df_dripLLAI[df_dripLLAI['Year'] == y]
    save.to_csv(name+'/drip_line_leaf_area_index'+str(y)+'.csv', index=False)
    save = df_CWNDVI[df_CWNDVI['Year'] == y]
    save.to_csv(name+'/crops_weeds_NDVI_full_sun_conditions'+str(y)+'.csv', index=False)
    save = df_LaicwAntPlotScale[df_LaicwAntPlotScale['Year'] == y]
    save.to_csv(name+'/leaf_area_index_crops_weeds_at_the_antenna_plot_scale'+str(y)+'.csv', index=False)
    save = df_LaiTreecwAntPlotScale[df_LaiTreecwAntPlotScale['Year'] == y]
    save.to_csv(name+'/leaf_area_index_of_tree_crops_weeds_at_the_antenna_plot_scale'+str(y)+'.csv', index=False)
    save = df_woodW[df_woodW['Year'] == y]
    save.to_csv(name+'/wood_weather'+str(y)+'.csv', index=False)
    save = df_netRadiation[df_netRadiation['Year'] == y]
    save.to_csv(name+'/net_radiation_at_30m_high'+str(y)+'.csv', index=False)

    # meteo
    save = df_msSmallAnt[df_msSmallAnt['Year'] == y]
    save.to_csv(name+'/meteo_station_small_antenna_at_2m_high'+str(y)+'.csv', index=False)
    save = df_msTallAnt[df_msTallAnt['Year'] == y]
    save.to_csv(name+'/meteo_station_tall_antenna_at_20m_high'+str(y)+'.csv', index=False)
    save = df_fluxTallAnt[df_fluxTallAnt['Year'] == y]
    save.to_csv(name+'/fluxes_tall_antenna_at_20m_high'+str(y)+'.csv', index=False)
```
